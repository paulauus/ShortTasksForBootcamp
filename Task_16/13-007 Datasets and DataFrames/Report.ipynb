```python
# Import pandas
import pandas as pd
```

```python
# Create a DataFrame with balance.txt
df = pd.read_csv("balance.txt", sep=" ")
df.head()
```

```python
# Compare the average income based on ethnicity.
avg_income_ethnicity = df.groupby("Ethnicity")["Income"].mean()
print(avg_income_ethnicity)
```

```python
# On average, do married or single people have a higher balance?
married_single_balance = df.groupby("Married")["Balance"].sum()
print(married_single_balance)
print("Married people have a higher balance.")
```

```python
# What is the highest income in our dataset?
highest_income = df["Income"].max()
print(f"The highest income in the dataset is {highest_income}.")
```

```python
# What is the lowest income in our dataset?
lowest_income = df["Income"].min()
print(f"The lowest income in the dataset is {lowest_income}.")
```

```python
# How many cards do we have recorded in our dataset?
cards_sum = df["Cards"].sum()
print(f"There are {cards_sum} cards recorded in the dataset.")
```

```python
# How many females do we have information for vs how many males?
df["Gender"].value_counts()
```
